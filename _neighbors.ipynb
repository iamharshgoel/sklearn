```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, y = make_blobs(n_features=400, centers=4, cluster_std=0.70, random_state=13)
```

```python
X.shape
```

```python
X[:5]
```

```python
y[:100]
```

```python
np.unique(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
plt.scatter(X_train[:,0], X_train[:,1])
plt.title("Four well behaved clusters")
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred.shape
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred, cmap='seismic',alpha=0.2)
```

```python
model.cluster_centers_
```

```python
ctrs = model.cluster_centers_
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred, cmap='seismic',alpha=0.2)
plt.scatter(ctrs[:,0],ctrs[:,1],c='white', edgecolors='black',s=150)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred, cmap='seismic',alpha=0.5)
plt.scatter(X_train[:,0],X_train[:,1],c=y_train, cmap='seismic',alpha=0.2)
```

```python
from sklearn.datasets.samples_generator import make_blobs

cluster_std = 0.7     # std of 0.70 gives tight clusters, try other options!
random_state = 13

X, y = make_blobs(n_samples=400,
                  centers=4,
                  cluster_std=cluster_std,
                  random_state=random_state)

X_train, X_test, y_train, y_test = train_test_split(X, y)

plt.scatter(X_train[:, 0], X_train[:, 1],
            c=y_train,
            cmap='seismic', alpha=0.5);
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
model.fit(X_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X_test[:, 0], X_test[:, 1],
            c=y_pred,
            cmap='seismic', alpha=0.5);
```
